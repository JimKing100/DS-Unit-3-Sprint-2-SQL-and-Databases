```
import sqlite3
```

```
!wget https://github.com/kts/sqlite-samples/blob/master/northwind.sqlite?raw=true -O northwind.sqlite
```

```
conn = sqlite3.connect('northwind.sqlite')
```

```
cur = conn.cursor()
```

```
query = """
SELECT OrderID, OrderDate, ShippedDate, CustomerID, Freight
FROM Orders
ORDER BY Freight DESC
"""
cur.execute(query)
cur.fetchall()
```

```
cur.fetchone()
```

```
[desc[0] for desc in cur.description]
```

```
import pandas as pd
```

```
df = pd.read_sql_query(query, conn)
```

```
df
```

```
query = """
SELECT o.OrderID, o.OrderDate, e.FirstName, e.LastName, od.ProductID
FROM Orders o, Employees e, "Order Details" od
ON o.EmployeeID = e.EmployeeID AND o.OrderID = od.OrderID
"""

df_2 = pd.read_sql(query, conn)
df_2
```

```
conn_new = sqlite3.connect('new_db.sqlite')
```

```
df_2.to_sql('my_table', conn)
```
